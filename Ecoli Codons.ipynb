```python
import wholecell.reconstruction.knowledge_base_ecoli as kbe
import numpy as np
import matplotlib.pyplot as plt
import Bio.Seq
%matplotlib inline
```

```python
# Loading knowledge base
kb = kbe.KnowledgeBaseEcoli(False)
```

### Codon frequency

```python
# Creating set of all codons in our sequences
all_codons = []
for sequence in [x['codingRnaSeq'] for x in kb._proteins]:
    n = len(sequence)
    for i in range(0, n - n%3, 3):
        codon = sequence[i:i+3]
        all_codons.append(codon)
unique_codons_set = set(all_codons)
unique_codons_list = list(unique_codons_set)
all_codons_index = [unique_codons_list.index(x) for x in all_codons]
```

```python
# Calculate codon frequency
codon_frequency = np.bincount(all_codons_index)
codon_frequency = np.sort(codon_frequency)
frequency_sorted_codon = [x for (y,x) in sorted(zip(codon_frequency,unique_codons_list))]
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
bars = plt.bar(range(len(codon_frequency)), codon_frequency)
xtickidx = ax.set_xticks(np.arange(0.5, len(frequency_sorted_codon) + 0.5, 1))
xticktxt = ax.set_xticklabels(frequency_sorted_codon)
allxtick = plt.xticks(rotation = 90)
plt.xlabel('codon')
plt.ylabel('frequency')
```

### tRNA codons available - No wobble

```python
# Import tRNA anti-codons from Ecocyc
from wholecell.reconstruction.knowledge_base_ecoli import TRNA_ANTICODON
unique_anti_codons = set([TRNA_ANTICODON[x] for x in TRNA_ANTICODON.keys()])
```

```python
# Find available codons without any wobble considerations
trna_codons = set([Bio.Seq.Seq(x).reverse_complement().tostring().replace('T','U') for x in unique_anti_codons])
trna_codons.update(['UAG','UAA','UGA']) # Add stop codons
```

```python
# Find coverage of our Ecoli codons
missing_codons = unique_codons_set.difference(trna_codons)
colors = ['b']*len(frequency_sorted_codon)
for x in missing_codons:
    colors[frequency_sorted_codon.index(x)] = 'r'
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
bars = plt.bar(range(len(codon_frequency)), codon_frequency, color = colors)
xtickidx = ax.set_xticks(np.arange(0.5, len(frequency_sorted_codon) + 0.5, 1))
xticktxt = ax.set_xticklabels(frequency_sorted_codon)
allxtick = plt.xticks(rotation = 90)
plt.xlabel('codon')
plt.ylabel('frequency')
```

```python
# Calculate missing coverage
len(missing_codons) / float(len(unique_codons_set))
```

### tRNA codons available - Watson and Crick wobble

```python
# Import tRNA anti-codons from Ecocyc
from wholecell.reconstruction.knowledge_base_ecoli import TRNA_ANTICODON
unique_anti_codons = set([TRNA_ANTICODON[x] for x in TRNA_ANTICODON.keys()])
```

```python
# Define wobble dict - 5' tRNA anti-codon to corresonding 3' RNA codon.
wobble_dict = {'A' : ['U'],
               'C' : ['G'],
               'G' : ['C','U'],
               'U' : ['A','G']}
```

```python
trna_codons = []
for anticodon in unique_anti_codons:
    new_codon_base = anticodon[1:]
    wobbles = wobble_dict[anticodon[0]]
    trna_codons.append(Bio.Seq.Seq(anticodon).reverse_complement().tostring().replace('T','U'))
    for w in wobbles:
        trna_codons.append(Bio.Seq.Seq(new_codon_base).reverse_complement().tostring().replace('T','U') + w)
trna_codons = set(trna_codons)
trna_codons.update(['UAG','UAA','UGA']) # Add stop codons
```

```python
# Find coverage of our Ecoli codons
missing_codons = unique_codons_set.difference(trna_codons)
colors = ['b']*len(frequency_sorted_codon)
for x in missing_codons:
    colors[frequency_sorted_codon.index(x)] = 'r'
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
bars = plt.bar(range(len(codon_frequency)), codon_frequency, color = colors)
xtickidx = ax.set_xticks(np.arange(0.5, len(frequency_sorted_codon) + 0.5, 1))
xticktxt = ax.set_xticklabels(frequency_sorted_codon)
allxtick = plt.xticks(rotation = 90)
plt.xlabel('codon')
plt.ylabel('frequency')
```

```python
# Calculate missing coverage
len(missing_codons) / float(len(unique_codons_set))
```

### tRNA mapping to codons

```python
from wholecell.reconstruction.knowledge_base_ecoli import TRNA_ANTICODON

# Create tRNA --> codon mapping dictionary
wobble_dict = {'A' : ['U'],
               'C' : ['G'],
               'G' : ['C','U'],
               'U' : ['A','G']}

trna_codon_association = {}
for trna in TRNA_ANTICODON.iterkeys():
    trna_codon_association[trna] = []
    anticodon = TRNA_ANTICODON[trna]
    new_codon_base = anticodon[1:]
    wobbles = wobble_dict[anticodon[0]]
    trna_codon_association[trna].append(Bio.Seq.Seq(anticodon).reverse_complement().tostring().replace('T','U'))
    for w in wobbles:
        trna_codon_association[trna].append(Bio.Seq.Seq(new_codon_base).reverse_complement().tostring().replace('T','U') + w)
    trna_codon_association[trna] = set(trna_codon_association[trna])
    
# Add non-standard wobble associations
for trna in trna_codon_association.keys():
    if 'CAU' == TRNA_ANTICODON[trna]:
        trna_codon_association[trna].update(['AUA'])
    if 'CCG' == TRNA_ANTICODON[trna]:
        trna_codon_association[trna].update(['CGA'])
    if 'ACG' == TRNA_ANTICODON[trna]:
        trna_codon_association[trna].update(['CGC'])

# Invert dictionary to get codon --> tRNA association
codon_trna_association = {}
for trna,codon_list in trna_codon_association.items():
    for codon in codon_list:
        if not codon_trna_association.has_key(codon):
            codon_trna_association[codon] = set([])
        codon_trna_association[codon].update([trna])
```

```python
codon_trna_association
```

```python
# Organize data
trnas = trna_codon_association.keys()
codons = codon_trna_association.keys()

import matplotlib
fig, ax = plt.subplots(figsize=(15, 15))
y = np.arange(len(trnas)+1)
x = np.arange(len(codons)+1)
#z = np.random.random((len(trnas) + 1, len(codons)+1))
z = np.zeros((len(trnas) + 1, len(codons)+1))

# Fill in association data
for trna in trnas:
    associated_codons = trna_codon_association[trna]
    for codon in associated_codons:
        z[trnas.index(trna), codons.index(codon)] = 1.

# Plot
plt.pcolor(x, y, z,cmap = 'binary')

# Set yticks to trnas
ytickidx = ax.set_yticks(np.arange(0.5, len(trna_codon_association.keys()) + 0.5, 1))
yticktxt = ax.set_yticklabels(trna_codon_association.keys())

# Set xticks to codons
xtickidx = ax.set_xticks(np.arange(0.5, len(codon_trna_association.keys()) + 0.5, 1))
xticktxt = ax.set_xticklabels(codon_trna_association.keys())
allxtick = plt.xticks(rotation = 90)

# Add labels
plt.xlabel('codon')
plt.ylabel('trna')
```
