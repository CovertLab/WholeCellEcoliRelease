Why partition <code>MoleculeCounts</code> state?
================================================
<code>MoleculeCounts</code> stores the counts of every molecule in the simulation.  The attribute <code>MoleculeCounts.counts</code> which is a numpy array with rows = the number of molecular species and columns = the number of compartments.  This is a sparse matrix so even though it looks like many zeros there is data in here! In this case there are 21,529 molecular species and 10 compartments.

```python
moleculeCounts = sim.getState("MoleculeCounts")
moleculeCounts.counts.shape
```

```python
moleculeCounts.counts
```

```python
# The molecular ID for each row
moleculeCounts.ids
```

```python
# The compartment ID for each column
moleculeCounts.compartments
```

All of the processes in the simulation need to share the same set of resources represented by counts in <code>MoleculeCounts</code>. In order to assure that the processes can be run in parallel, and that we don't violate any physical laws while doing so (i.e. mass conservation), we choose to "partition" each state among the processes.

Partitioning accomplishes two goals:
<ol>
<li>Each process will be allocated resouces proportional to its required need such that mass is conserved and there are no conflicts in usage of resources. </li>
<li>Each process will only see the minimal necessary rows of the whole MoleculeCounts state required for its functions. This simplifies computation and coding implementation for the process. You won't need to go searching through large number of indicies in MoleculeCounts to find the molecular species required for modeling, only the limited number of rows in the processes' partition.</li>
</ol>

What is a partition?
====================
If you look at the paritions that exist in molecule counts <code>moleculeCounts.partitions</code> the result is a list of all partitions:

```python
moleculeCounts.partitions
```

```python
# Names for each parition
pprint.pprint([x.meta['id'] for x in moleculeCounts.partitions])
```

Note that there can be multiple partitions per process. This is purely for organizational purposes within the process. For example, the Transcription process requires metabolites (NTPs, ATP, etc.), enzymes (RNA polymerase), and RNAs for it to polymerize. These 3 requirements correspond to its 3 partitions.

What is a partition? An exact copy of <code>MoleculeCounts</code> state with a few small changes and added properties:
<ol>
<li>The rows in <code>partition.counts</code> only correspond to the molecular species being divided up by that partition. For example, the <code>Transcription.metabolites</code> partition will only include rows for ATP[c], CTP[c], GTP[c], UTP[c], PPI[c], H2O[c], and H[c]. Each partition still includes all compartment columns.</li>
<li>A pointer to a function in the process which calculates a request or estimate from the process of how many of each molecular species the process will require during the next timestep. For example, in the <code>Transcription</code> process <code>Transcription.calcReqMetabolites()</code> is stored in <code>partition.reqFunc</code>.</li>
<li>A few additional properties like <code>partition.partitionedCounts</code> are set in order to do the book-keeping for partitioning. These properties are set to None in the state.</li>
</ol>

```python
# Get all partitions associated with transcription
transcription_partitions = [x for x in moleculeCounts.partitions if x.meta['id'] == 'MoleculeCounts_Transcription']
# Pick out the one which allocates metabolites
metabolite_partition = [x for x in transcription_partitions if 'ATP' in x.ids][0]
```

Partitioning algorithm
======================
The <code>evolveState</code> function in the simulation has the following code:<br>
<table><tr><td><code>def evolveState(self):</code>
<code>
    # Prepare to partition states among processes
    for state in self.states:
        state.prepartition()

    # Partition states among processes
    for state in self.states:
        state.partition()

    # Simulate submodels
    for process in self.processes:
        process.evolveState()

    # Merge state
    for state in self.states:
        state.merge()

    # Recalculate dependent state
    for state in self.states:
        state.calculate()
</code>
</td></tr></table>

We will go through <code>state.prepartition()</code>, <code>state.partition()</code>, and <code>state.merge()</code> here.

Prepartitioning
---------------
The default behavior of the prepartition function is to do nothing.
<table><tr><td><code>def prepartition(self):</code>
<code>
    return
</code>
</td></tr></table>
This function can be defined for certain states. For example, in <code>MoleculeCounts</code> <code>partition.fullCounts</code> values are set to the corresponding rows values in the state:
<table><tr><td><code>def prepartition(self):</code>
<code>
    for partition in self.partitions:
			partition.fullCounts = self.counts[numpy.unravel_index(partition.mapping, self.counts.shape)]
</code>
</td></tr></table>
This information isn't used int he <code>MoleculeCounts.partition()</code> function but it is used later in the <code>reqFunc</code>.

Partitioning
------------
The default behavior of the parition function is to copy the entire state:
<table><tr><td><code>def partition(self):</code>
<code>
    for partition in self.partitions:
			for dynamic in self.meta["dynamics"]:
				setattr(partition, dynamic, getattr(self, dynamic))
</code>
</td></tr></table>
This will only occur in states where multiple processes do not simultaneously edit state properties.

From now on we will focus on the <code>MoleculeCounts</code> partition function.
