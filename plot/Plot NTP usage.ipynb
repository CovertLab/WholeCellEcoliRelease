```python
# Import packages
import tables
import numpy
import os
import matplotlib.pyplot as plt
```

```python
# Load data (slow)
FIXTURE_DIR = os.path.join('out', 'test', 'Test_Transcription')

dirs = os.walk(FIXTURE_DIR).next()[1]
nSims = len(dirs)
simLength = 500

ntpIDs = ['ATP', 'UTP', 'CTP', 'GTP']

assignedIdxs = False

ntpIdxs = None

startAt = 2 # 0 has no production, and 1 is prior to the model receiving excess NTPs (fix?)
stride = 10

timePoints = numpy.arange(startAt, simLength+1, stride).size

ntpUsage = numpy.zeros((timePoints, len(ntpIDs), nSims), float)

for iSim, simDir in enumerate(dirs):
    with tables.openFile(os.path.join(FIXTURE_DIR, simDir, 'MoleculeCounts.hdf')) as h5file:
        if not assignedIdxs:
            names = h5file.get_node('/names')
            
            molIDs = names.molIDs.read()
            processes = names.processes.read()
            compartments = names.compartments.read()
            
            ntpIdxs = numpy.array([molIDs.index(id_) for id_ in ntpIDs])
            processIdx = processes.index('Transcription')
            compartmentIdx = compartments.index('c')
            
            assignedIdxs = True
        
        mc = h5file.root.MoleculeCounts

        ntpUsage[..., iSim] = (mc.read(startAt, None, stride, 'countsBulkPartitioned')[:, ntpIdxs, compartmentIdx, processIdx]
                               - mc.read(startAt, None, stride, 'countsBulkReturned')[:, ntpIdxs, compartmentIdx, processIdx])
    
```

```python
# Plot average NTP usage
time = numpy.arange(startAt, simLength+1, stride)/60. # TODO: load sim length

ntpUsageAvg = ntpUsage.mean(2)/1000
ntpUsageMax = ntpUsage.max(2)/1000
ntpUsageMin = ntpUsage.min(2)/1000

colors = ['green', 'blue', 'orange', 'red']

plots = []

for i, color in enumerate(colors):
    plt.fill_between(time, ntpUsageMin[:, i], ntpUsageMax[:, i], alpha = 0.5, linewidth = 0, facecolor = color)
    plots.append( plt.plot(time, ntpUsageAvg[:, i], color = color)[0] )

plt.title('Average NTP usage by Transcription')
plt.legend(plots, ntpIDs, loc = 0)
plt.xlabel('Time (minutes)')
plt.ylabel('Molecule count (thousands)')
```

```python
# Plot actual vs. expected NTP usage
expectedNormalized = numpy.exp(numpy.log(2)/60. * time)
expectedNormalized /= expectedNormalized.mean()

for i, ntpID in enumerate(ntpIDs):
    plt.subplot(2, 2, i+1)
    plt.plot(time, ntpUsageAvg[:, i], color = colors[i], marker = '.', linestyle = 'None')
    plt.plot(time, expectedNormalized * ntpUsageAvg[:, i].mean(), color = 'black')
    #plt.title(ntpID)

# I think 500s is too short of a time to reasonably show an exponential growth trend for a 3600s doubling time - JM
```
